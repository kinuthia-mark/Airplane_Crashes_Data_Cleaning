```python
movies_2021 = df[(df['type'] == 'Movie') & (df['release_year'] == 2021)]
print(f"\n4. Cast of all movires released in 2021:")
for idx,row in movies_2021.iterrows():
    print(f"\n{row['title']}:")
    print(f"Cast:{row['cast']}:")
    
```
